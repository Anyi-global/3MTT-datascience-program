## Data Visualization with Matplotlib

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load dataset
df = pd.read_csv('AB_NYC_2019.csv')
```

```python
# Check Number of Missing Values
print(df.isnull().sum())
```

```python
# Check Number of Duplicate Values
print(df.duplicated().sum())
```

```python
## Handle Missing Values by filling with the average
# fill specific columns with their average using fillna().mean() function
df["reviews_per_month"].fillna(df["reviews_per_month"].mean(), inplace=True)
df["price"].fillna(df["price"].mean(), inplace=True)
df["number_of_reviews"].fillna(df["number_of_reviews"].mean(), inplace=True)
```

```python
# Drop rows where data is still missing and duplicate
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
```

## Data Visualizations

```python
# Plot a bar chart of the prices
df["price"].plot()
plt.ylabel("Price")
plt.xlabel("Price Categories")
plt.title("Plot of Prices")
plt.show()

# Note: Bar chart is the default chart if none is specified
```

```python
# Scatter plot of price and number_of_reviews
df.plot(kind="scatter", x="price", y="number_of_reviews", color="blue")
plt.title("Price vs Number of Reviews")
plt.show()
```

## Subplots

```python
fig, axis = plt.subplots(2, 3, figsize=(15,10))

# Price Vs Frequency
axis[0,0].hist(df['price'], bins=20, color='lightblue')
axis[0,0].set_title("Price vs Frequency")
axis[0,0].set_xlabel("Price")
axis[0,0].set_ylabel("Frequency")

# Number of Reviews Vs Frequency
axis[0,1].hist(df['number_of_reviews'], bins=20, color='red')
axis[0,1].set_title("Number of Reviews vs Frequency")
axis[0,1].set_xlabel("Number of Reviews")
axis[0,1].set_ylabel("Price")

# Group by type of room and average price
room_type = df.groupby('room_type')['price'].mean().reset_index()

axis[0,2].bar(room_type['room_type'], room_type['price'], color='lightgreen')
axis[0,2].set_title("Average Price by Room Type")
axis[0,2].set_xlabel("Room Type")
axis[0,2].set_ylabel("Price")

# Number of listings in each borough (NYC)
neighbourhoods = df['neighbourhood_group'].value_counts().reset_index()
neighbourhoods.columns = ['neighbourhood_group', 'count']

axis[1,0].bar(neighbourhoods['neighbourhood_group'], neighbourhoods['count'], color='orange')
axis[1,0].set_title("Listings by Neighbourhood")
axis[1,0].set_xlabel("Neighbourhood")
axis[1,0].set_ylabel("Number of Listings")

# Price vs Number of Reviews
axis[1,1].bar(df['price'], df['number_of_reviews'], color='blue')
axis[1,1].set_title("Price vs Reviews")
axis[1,1].set_xlabel("Price")
axis[1,1].set_ylabel("Number of Reviews")

# Host names vs Amount Paid (price)
grouped_host = df.groupby('host_name')
axis[1,2].bar(grouped_host, df['price'], color='purple')
axis[1,2].set_title("Amount Paid by Each Host")
axis[1,2].set_xlabel("Host Name")
axis[1,2].set_ylabel("Price")

plt.tight_layout()
plt.show()
```
