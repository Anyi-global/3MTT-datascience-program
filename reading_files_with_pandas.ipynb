```python
del df['Unnamed: 2']  #deleting a column in the dataframe
```

```python
df.groupby(['Region']).count()  #categorizing the Region column using groupby() method
```

```python
df['Country']  #selecting only the Country column from the dataframe
```

```python
df.loc[224, 'Country']  #Label-based indexing. You use row labels and column labels to select data. Includes the end value in slicing.

#Works with boolean arrays or index names.
```

```python
df.loc[100:105, ['Country', 'Region']]  # Get rows with labels 100 to 105, and columns 'Country' and 'Region'
```

```python
df.iloc[220]  # Integer-location based indexing. Only works with integers. You use row and column positions (integers).
```

```python
df.iloc[101, 0]  # Get the value at the 101 row and 1st column (index 1, 0)
```

```python
df.iloc[100:104, 0:2]  # Get rows 100 to 103 and columns 0 to 1
```
