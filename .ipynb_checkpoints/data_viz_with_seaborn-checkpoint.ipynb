## Univariate non visual statistical analysis

Univariate analysis refers to one-variable analysis

```python
#splitting the data into two distinct groups
#cat_df = categorical columns, categorical refers to object data types which are the same thing as strings
#num_df = numerical columns

cat_df = df.select_dtypes(include=['object']) #returns dataframe that contain the object data type
num_df = df.select_dtypes(include=['number']) #returns columns that contain the numerical data type

cat_df.head()
```

```python
num_df.head()
```

```python
#creating functions

#discrete = categorical
#univariate = one

def discrete_uni_analysis(data):
    for col_name in data:
        print('\n','*'*5, col_name, '*'*5) #for formatting
        print(data[col_name].agg(['count', 'nunique', 'unique']))
        print('Value_counts: \n', data[col_name].value_counts())

#count the number of the rows of data
#nunique = the number of unique items in the column
#unique = find out unique items
#value counts = for each unique item in our column, how many times do they appear

def num_uni_analysis(data):
    for col_name in data:
        print('\n','*'*5, col_name, '*'*5)
        print(data[col_name].agg(['count', 'min', 'max', 'mean','skew', 'median', 'std']))
```

```python
#using the functions
discrete_uni_analysis(cat_df)
```

```python
num_uni_analysis(num_df)
```

## Univariate plotting for num columns - histogram, KDE, ECDF and box plot

## ðŸ“Š Histogram

A histogram shows how data is spread out.

- Imagine you collect everyone's height in a class.
- You group the heights into ranges (like 150â€“155 cm, 155â€“160 cm, etc.).
- The histogram shows how many people fall into each range using bars.

ðŸŸ¦ Each bar = number of data points in that range
âœ… Good for seeing shape of the data (is it skewed? symmetric? etc.)

## ðŸ“ˆ KDE (Kernel Density Estimate)

A KDE is a smoothed version of a histogram.

- Instead of bars, it draws a smooth curve to show the shape of the data distribution.

- It's like putting a soft, fuzzy curve over the histogram to smooth out the noise.

âœ… Great for comparing distributions
âœ… Shows where the data is concentrated

## ðŸ“‰ ECDF (Empirical Cumulative Distribution Function)

An ECDF shows the percent of data that's less than or equal to a certain value.

- Say you want to know how many people are shorter than 160 cm.

- An ECDF line tells you exactly that â€” for any value.

ðŸ”¹ X-axis = value (e.g., height)
ðŸ”¹ Y-axis = proportion of data â‰¤ that value

âœ… Helps you see percentiles, medians, etc. easily
âœ… More precise than a histogram for individual values

## ðŸ“¦ Box Plot (Box-and-Whisker Plot)

A box plot shows the summary of a dataset.

- The box shows the middle 50% (called the interquartile range)

- The line inside the box is the median

- The whiskers stretch to show the general range (not counting outliers)

- Dots beyond the whiskers are outliers

âœ… Great for comparing multiple groups
âœ… Quickly shows spread, center, and outliers



```python
# to see the distribution of the total bill in the data
sns.histplot(data=df, x='total_bill', bins=20)

plt.show()
```

```python
# using the boxplot
sns.boxplot(data=df, x='total_bill');
```

```python
# using the kde plot
sns.kdeplot(data=df, x='total_bill');
```

```python
# using an ecdf, it visualizes all the data points but is less intuitive
sns.ecdfplot(data=df, x='total_bill');
```

```python
# presenting all the plots using subplots
fig, axs = plt.subplots(1,5, figsize=(10,4), layout='constrained')
plt.suptitle('Numerical features visualization with seaborn')

axs[0].set_title('Histogram')
sns.histplot(data=df, x='total_bill', ax=axs[0])

axs[1].set_title('KDE')
sns.kdeplot(data=df, x='total_bill', ax=axs[1])

axs[2].set_title('ECDF')
sns.ecdfplot(data=df, x='total_bill', ax=axs[2])

axs[3].set_title('Box plot')
sns.boxplot(data=df, x='total_bill', ax=axs[3])

axs[4].set_title('Boxen plot')
sns.boxenplot(data=df, x='total_bill', ax=axs[4]);
```

## Bivariate plotting for "num vs cat" columns

```python
sns.histplot(data=df, x='total_bill', hue='day', alpha=0.7)
```

```python
#using subplots

fig, axs = plt.subplots(1,4, figsize=(10,3), constrained_layout=True)
plt.suptitle('Num vs. Cat features')

axs[0].set_title('Box plot')
sns.boxplot(data=df, x='day', y='total_bill', ax=axs[0])

axs[1].set_title('Violin plot')
sns.violinplot(data=df, x='day', y='total_bill', ax=axs[1])

axs[2].set_title('Boxen plot')
sns.boxenplot(data=df, x='day', y='total_bill', ax=axs[2])

axs[3].set_title('Histogram plot')
sns.histplot(data=df.loc[df['day']=='Sat'], x='total_bill', alpha=0.3, ax=axs[3], label='Sat')
sns.histplot(data=df.loc[df['day']=='Fri'], x='total_bill', alpha=0.3, ax=axs[3], label='Fri')
axs[3].legend()
```

## Multiplot grid using FacetGrid

It allows you to create a grid of subplots (facets) based on the unique values of one or more categorical variables in your dataset. Each subplot displays the same type of plot but filtered to show data for a specific category or combination of categories.

### Question:
Which gender pays more?

```python
#without facetgrid

fig, axs = plt.subplots(1,2, figsize=(5,3), constrained_layout=True)
plt.suptitle('Which gender pays more?')

axs[0].set_title('Female')
sns.histplot(data=df.loc[df['sex']=='Female'], x='total_bill', ax=axs[0], label='Female')
axs[0].legend()

axs[1].set_title('Male')
sns.histplot(data=df.loc[df['sex']=='Male'], x='total_bill', ax=axs[1], label='Male')
axs[1].legend()
```

Notice how both plots have different scales for the y axis. This makes it difficult to comapre both plots.

```python
#using facetgrid

grid = sns.FacetGrid(df, col='sex')
grid.map(sns.histplot, 'total_bill')
```

```python
grid = sns.FacetGrid(df, col='sex')
grid.map(sns.histplot, 'tip')
```

```python
grid = sns.FacetGrid(df, col='sex', row='smoker')
grid.map(sns.histplot, 'tip')
```

```python
grid = sns.FacetGrid(df, col='sex', hue='smoker')
grid.map(sns.histplot, 'total_bill', alpha=0.3)
grid.add_legend()
```

```python
grid = sns.FacetGrid(df, col='sex', hue='smoker')
grid.map(sns.scatterplot, 'total_bill', 'tip', alpha=0.5)
grid.add_legend()
```

**Difference Between FacetGrid and PairGrid**

- FacetGrid: Plots the same kind of chart across subsets of data.
- PairGrid: Plots relationships between different pairs of variables in a grid format.
